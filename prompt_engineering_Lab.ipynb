# Data Extraction

```python
from openai import OpenAI
import re

client = OpenAI()

def call_openai(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt
    )
    return response.output_text.strip()

# Initial simple prompt for data extraction
extraction_prompt_v1 = """
Extract information from this customer feedback: "I ordered item #12345 on March 15th. The delivery was fast but the packaging was damaged."
"""

def analyze_extraction_output(text):
    normalized = text.lower()

    # Check for expected information
    has_order_number = "12345" in normalized
    has_order_date = any(date in normalized for date in ["march 15", "march 15th", "3/15", "03/15"])
    has_delivery_feedback = "fast" in normalized
    has_packaging_feedback = "damaged" in normalized

    # Basic format observations
    has_bullets = "-" in text or "•" in text
    has_labels = ":" in text

    completeness_score = sum([
        has_order_number,
        has_order_date,
        has_delivery_feedback,
        has_packaging_feedback
    ]) / 4

    print("\n" + "="*50)
    print("DATA EXTRACTION ANALYSIS")
    print("="*50)
    print(f"Order number found: {has_order_number}")
    print(f"Order date found: {has_order_date}")
    print(f"Delivery feedback found: {has_delivery_feedback}")
    print(f"Packaging feedback found: {has_packaging_feedback}")
    print(f"Uses bullets: {has_bullets}")
    print(f"Uses labels/structured format: {has_labels}")
    print(f"Completeness score: {completeness_score:.2%}")

    if completeness_score < 1.0:
        print("Some information may be missing.")
    else:
        print("All expected information appears to be present.")

# Run extraction prompt
result = call_openai(extraction_prompt_v1)

print("Data Extraction Result:")
print(result)

# Analyze extracted output
analyze_extraction_output(result)
```

```python
from openai import OpenAI
from collections import Counter
import re

client = OpenAI()

def call_openai(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt
    )
    return response.output_text.strip()

# Initial simple prompt for product description
product_prompt_v1 = """
Create a product description for a wireless mouse that costs $29.99.
"""

def normalize_text(text):
    """
    Normalize text for softer agreement comparison.
    - lowercase
    - strip leading/trailing spaces
    - collapse repeated whitespace
    """
    text = text.lower().strip()
    text = re.sub(r"\s+", " ", text)
    return text

def test_iterations(prompt, iteration_counts=[5, 10, 15]):
    ranked_results = []

    for n in iteration_counts:
        print(f"\n{'='*60}")
        print(f"Testing prompt with {n} iterations")
        print(f"{'='*60}")

        raw_outputs = []
        normalized_outputs = []

        for i in range(n):
            output = call_openai(prompt)
            raw_outputs.append(output)
            normalized_outputs.append(normalize_text(output))

            print(f"\n--- Run {i+1} ---")
            print(output)

        raw_counts = Counter(raw_outputs)
        normalized_counts = Counter(normalized_outputs)

        top_raw_output, top_raw_count = raw_counts.most_common(1)[0]
        top_norm_output, top_norm_count = normalized_counts.most_common(1)[0]

        exact_agreement_rate = top_raw_count / n
        normalized_agreement_rate = top_norm_count / n

        summary = {
            "iterations": n,
            "raw_outputs": raw_outputs,
            "normalized_outputs": normalized_outputs,
            "unique_raw_outputs": len(raw_counts),
            "unique_normalized_outputs": len(normalized_counts),
            "top_raw_count": top_raw_count,
            "top_normalized_count": top_norm_count,
            "exact_agreement_rate": exact_agreement_rate,
            "normalized_agreement_rate": normalized_agreement_rate,
            "most_common_raw_output": top_raw_output,
            "most_common_normalized_output": top_norm_output
        }

        ranked_results.append(summary)

        print("\nSummary:")
        print(f"Unique raw outputs: {len(raw_counts)}")
        print(f"Unique normalized outputs: {len(normalized_counts)}")
        print(f"Exact agreement rate: {exact_agreement_rate:.2%}")
        print(f"Normalized agreement rate: {normalized_agreement_rate:.2%}")

    # Rank by normalized agreement first, then exact agreement
    ranked_results = sorted(
        ranked_results,
        key=lambda x: (x["normalized_agreement_rate"], x["exact_agreement_rate"]),
        reverse=True
    )

    return ranked_results

def print_ranking(ranked_results):
    print(f"\n\n{'#'*60}")
    print("RANKING BY AGREEMENT")
    print(f"{'#'*60}")

    for rank, result in enumerate(ranked_results, start=1):
        print(f"\nRank #{rank}")
        print(f"Iterations: {result['iterations']}")
        print(f"Normalized agreement rate: {result['normalized_agreement_rate']:.2%}")
        print(f"Exact agreement rate: {result['exact_agreement_rate']:.2%}")
        print(f"Unique raw outputs: {result['unique_raw_outputs']}")
        print(f"Unique normalized outputs: {result['unique_normalized_outputs']}")

# Run the test
ranked_results = test_iterations(
    product_prompt_v1,
    iteration_counts=[5, 10, 15]
)

print_ranking(ranked_results)
```

```python
from openai import OpenAI
from collections import Counter
import re

client = OpenAI()

def call_openai(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt
    )
    return response.output_text.strip()

# Prompt to test
extraction_prompt_v1 = """
Extract information from this customer feedback:
"I ordered item #12345 on March 15th. The delivery was fast but the packaging was damaged."
"""

# Expected fields for completeness checking
expected_fields = {
    "order_number": ["12345"],
    "order_date": ["march 15", "march 15th", "3/15", "03/15"],
    "delivery_feedback": ["fast"],
    "packaging_feedback": ["damaged"]
}

def normalize_text(text):
    """Lowercase and collapse whitespace for easier comparison."""
    text = text.lower().strip()
    text = re.sub(r"\s+", " ", text)
    return text

def run_prompt(prompt, n=5):
    """Run a prompt n times and collect responses."""
    responses = []

    for i in range(n):
        response = call_openai(prompt)
        responses.append(response)

        print(f"\nRun {i+1}:")
        print(response)

    return responses

def analyze_consistency(responses):
    """Measure how often the model gives the same response."""
    normalized_responses = [normalize_text(r) for r in responses]
    counts = Counter(normalized_responses)

    most_common_response, most_common_count = counts.most_common(1)[0]
    agreement_rate = most_common_count / len(responses)

    return {
        "unique_responses": len(counts),
        "agreement_rate": agreement_rate,
        "response_distribution": counts,
        "most_common_response": most_common_response
    }

def check_response_completeness(response, expected_fields):
    """Check whether a response contains the expected information."""
    normalized_response = normalize_text(response)

    found_fields = []
    missing_fields = []

    for field, keywords in expected_fields.items():
        if any(keyword in normalized_response for keyword in keywords):
            found_fields.append(field)
        else:
            missing_fields.append(field)

    completeness_score = len(found_fields) / len(expected_fields)

    return {
        "found_fields": found_fields,
        "missing_fields": missing_fields,
        "completeness_score": completeness_score
    }

def analyze_completeness(responses, expected_fields):
    """Measure completeness across all responses."""
    results = []
    total_score = 0
    fully_complete = 0

    for i, response in enumerate(responses, 1):
        result = check_response_completeness(response, expected_fields)
        results.append(result)
        total_score += result["completeness_score"]

        if result["completeness_score"] == 1.0:
            fully_complete += 1

        print(f"\nRun {i} Completeness:")
        print(f"  Score: {result['completeness_score']:.2%}")
        print(f"  Found: {result['found_fields']}")
        print(f"  Missing: {result['missing_fields']}")

    average_completeness = total_score / len(responses)
    full_completeness_rate = fully_complete / len(responses)

    return {
        "average_completeness": average_completeness,
        "full_completeness_rate": full_completeness_rate,
        "fully_complete_count": fully_complete,
        "details": results
    }

def analyze_prompt(prompt, expected_fields, n=5):
    """Run the full prompt evaluation."""
    print("\n" + "=" * 60)
    print("RUNNING PROMPT TEST")
    print("=" * 60)

    responses = run_prompt(prompt, n=n)

    print("\n" + "=" * 60)
    print("CONSISTENCY ANALYSIS")
    print("=" * 60)
    consistency = analyze_consistency(responses)
    print(f"Total responses: {n}")
    print(f"Unique responses: {consistency['unique_responses']}")
    print(f"Agreement rate: {consistency['agreement_rate']:.2%}")

    print("\nResponse distribution:")
    for response, count in consistency["response_distribution"].items():
        print(f"- {count} time(s): {response}")

    print("\n" + "=" * 60)
    print("COMPLETENESS ANALYSIS")
    print("=" * 60)
    completeness = analyze_completeness(responses, expected_fields)

    print("\n" + "=" * 60)
    print("FINAL SUMMARY")
    print("=" * 60)
    print(f"Agreement rate: {consistency['agreement_rate']:.2%}")
    print(f"Average completeness: {completeness['average_completeness']:.2%}")
    print(f"Fully complete responses: {completeness['fully_complete_count']}/{n}")
    print(f"Full completeness rate: {completeness['full_completeness_rate']:.2%}")

    return {
        "responses": responses,
        "consistency": consistency,
        "completeness": completeness
    }

# Run the analysis
results = analyze_prompt(extraction_prompt_v1, expected_fields, n=5)
```

4 runs start with: “Here is the extracted information from the customer feedback:”

```python
from openai import OpenAI
from collections import Counter
import re

client = OpenAI()

def call_openai(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt
    )
    return response.output_text.strip()

# Sentiment prompt
sentiment_prompt_v1 = """
Classify this customer message: "I love this product! It's exactly what I needed."
Return only one word: Positive, Negative, or Neutral.
"""

def normalize_text(text):
    text = text.lower().strip()
    text = re.sub(r"\s+", " ", text)
    return text

def run_prompt_n_times(prompt, n=5):
    responses = []
    print(f"\n{'#'*60}")
    print(f"# RUNNING PROMPT {n} TIMES")
    print(f"{'#'*60}")
    
    for i in range(n):
        result = call_openai(prompt)
        responses.append(result)
        print(f"Run {i+1}: {result}")
    
    return responses

def analyze_consistency(responses, run_label=""):
    normalized = [normalize_text(r) for r in responses]
    counts = Counter(normalized)

    most_common_response, most_common_count = counts.most_common(1)[0]
    agreement_rate = most_common_count / len(responses)

    print("\n" + "="*50)
    print(f"CONSISTENCY ANALYSIS ({run_label})")
    print("="*50)
    print(f"Total responses: {len(responses)}")
    print(f"Unique responses: {len(counts)}")
    print(f"Most common response: {most_common_response}")
    print(f"Agreement rate: {agreement_rate:.2%}")

    print("\nResponse distribution:")
    for response, count in counts.items():
        print(f"- {response}: {count} time(s)")

    return {
        "run_label": run_label,
        "total": len(responses),
        "unique_responses": len(counts),
        "agreement_rate": agreement_rate,
        "most_common": most_common_response,
        "counts": counts
    }

def analyze_completeness(responses, run_label=""):
    valid_labels = {"positive", "negative", "neutral"}
    complete_count = 0
    incomplete_responses = []

    print("\n" + "="*50)
    print(f"COMPLETENESS ANALYSIS ({run_label})")
    print("="*50)

    for i, response in enumerate(responses, 1):
        normalized = normalize_text(response)

        if normalized in valid_labels:
            complete_count += 1
            print(f"Run {i}: Complete ({response})")
        else:
            incomplete_responses.append((i, response))
            print(f"Run {i}: Incomplete ({response})")

    completeness_rate = complete_count / len(responses)
    print(f"\nCompleteness rate: {completeness_rate:.2%}")

    if incomplete_responses:
        print("\nIncomplete responses:")
        for run_num, response in incomplete_responses:
            print(f"- Run {run_num}: {response}")

    return {
        "run_label": run_label,
        "completeness_rate": completeness_rate,
        "complete_count": complete_count,
        "incomplete_responses": incomplete_responses
    }

def compare_runs(results_list):
    """Compare results across different run sizes."""
    print("\n" + "#"*60)
    print("# COMPARISON ACROSS RUN SIZES (5 vs 10 vs 15)")
    print("#"*60)

    print(f"\n{'Metric':<30}{'5 runs':<15}{'10 runs':<15}{'15 runs':<15}")
    print("-" * 75)

    # Consistency comparison
    cons_5, cons_10, cons_15 = results_list[0]["consistency"], results_list[1]["consistency"], results_list[2]["consistency"]
    comp_5, comp_10, comp_15 = results_list[0]["completeness"], results_list[1]["completeness"], results_list[2]["completeness"]

    print(f"{'Unique responses':<30}{cons_5['unique_responses']:<15}{cons_10['unique_responses']:<15}{cons_15['unique_responses']:<15}")
    print(f"{'Agreement rate':<30}{cons_5['agreement_rate']:<15.2%}{cons_10['agreement_rate']:<15.2%}{cons_15['agreement_rate']:<15.2%}")
    print(f"{'Completeness rate':<30}{comp_5['completeness_rate']:<15.2%}{comp_10['completeness_rate']:<15.2%}{comp_15['completeness_rate']:<15.2%}")
    print(f"{'Incomplete count':<30}{len(comp_5['incomplete_responses']):<15}{len(comp_10['incomplete_responses']):<15}{len(comp_15['incomplete_responses']):<15}")

    # Failure pattern detection
    print("\n" + "="*60)
    print("FAILURE PATTERN DETECTION")
    print("="*60)

    unique_5 = set(cons_5["counts"].keys())
    unique_10 = set(cons_10["counts"].keys())
    unique_15 = set(cons_15["counts"].keys())

    new_in_10 = unique_10 - unique_5
    new_in_15 = unique_15 - unique_10 - unique_5

    if new_in_10:
        print(f"\n⚠️  New response variations revealed at 10 runs (not seen in 5 runs):")
        for r in new_in_10:
            print(f"   - '{r}'")
    else:
        print("\n✅ No new variations revealed at 10 runs.")

    if new_in_15:
        print(f"\n⚠️  New response variations revealed at 15 runs (not seen in 5 or 10 runs):")
        for r in new_in_15:
            print(f"   - '{r}'")
    else:
        print("\n✅ No new variations revealed at 15 runs.")

    # Stability assessment
    print("\n" + "="*60)
    print("STABILITY ASSESSMENT")
    print("="*60)
    agreement_drop = cons_5["agreement_rate"] - cons_15["agreement_rate"]
    if abs(agreement_drop) < 0.05:
        print(f"✅ Agreement rate is STABLE across run sizes (Δ = {agreement_drop:+.2%}).")
    elif agreement_drop > 0:
        print(f"⚠️  Agreement rate DROPPED with more runs (Δ = {agreement_drop:+.2%}).")
        print("   → 5-run test may have OVERESTIMATED consistency.")
    else:
        print(f"📈 Agreement rate IMPROVED with more runs (Δ = {agreement_drop:+.2%}).")
        print("   → 5-run test may have UNDERESTIMATED consistency.")


# === EXECUTE TESTS ===
all_results = []

for n in [5, 10, 15]:
    responses = run_prompt_n_times(sentiment_prompt_v1, n=n)
    consistency = analyze_consistency(responses, run_label=f"{n} runs")
    completeness = analyze_completeness(responses, run_label=f"{n} runs")
    all_results.append({
        "n": n,
        "responses": responses,
        "consistency": consistency,
        "completeness": completeness
    })

# Compare across all run sizes
compare_runs(all_results)
```

Observation:

This result shows that the prompt is highly stable and consistent. Across 5, 10, and 15 runs, the model returned the exact same response — “Positive” — every time. That means there was only 1 unique response, giving an agreement rate of 100% and a completeness rate of 100%.

In simple terms, the prompt is working very well because the model is not changing its wording or output format between runs. No new variations appeared as the number of runs increased, so the behavior stayed reliable even at 15 runs. This means the prompt easily meets the project requirement of more than 80% consistency, and there are no visible failure patterns in this test.

```python



from openai import OpenAI
from collections import Counter
import re
import json

client = OpenAI()

def call_openai(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt
    )
    return response.output_text.strip()

# Version 1: simple prompt
extraction_prompt_v1 = """
Extract information from this customer feedback: "I ordered item #12345 on March 15th. The delivery was fast but the packaging was damaged."
"""

# Version 2: improved prompt
extraction_prompt_v2 = """
You are a strict data extraction assistant.

Task:
Extract structured information from the customer feedback below.

Output format:
Return the result ONLY as valid JSON with these exact keys:
{
  "order_id": "",
  "order_date": "",
  "delivery_status": "",
  "packaging_status": "",
  "overall_sentiment": ""
}

Rules:
- Use only information explicitly stated in the feedback.
- If a field is not mentioned, set its value to "N/A".
- Do not invent, guess, or infer details that are not in the text.
- Do not add explanations, comments, or extra text.
- Do not include markdown or code fences.
- Keep each value short (max 10 words).
- "overall_sentiment" must be exactly one of: Positive, Negative, Neutral.

Examples:

Feedback: "My order #99887 arrived on time and in perfect condition."
Output:
{
  "order_id": "99887",
  "order_date": "N/A",
  "delivery_status": "On time",
  "packaging_status": "Perfect condition",
  "overall_sentiment": "Positive"
}

Feedback: "Item #55501 was ordered on April 2nd. It never arrived."
Output:
{
  "order_id": "55501",
  "order_date": "April 2nd",
  "delivery_status": "Not delivered",
  "packaging_status": "N/A",
  "overall_sentiment": "Negative"
}

Now extract from this feedback:
"I ordered item #12345 on March 15th. The delivery was fast but the packaging was damaged."
"""

def normalize_text(text):
    text = text.lower().strip()
    text = re.sub(r"\s+", " ", text)
    return text

def is_valid_json(text):
    try:
        json.loads(text)
        return True
    except:
        return False

def test_prompt(prompt, label, iteration_counts=[5, 10, 15]):
    all_results = []

    print(f"\n{'#'*70}")
    print(f"TESTING {label}")
    print(f"{'#'*70}")

    for n in iteration_counts:
        print(f"\n{'='*60}")
        print(f"RUNNING {label} {n} TIMES")
        print(f"{'='*60}")

        raw_outputs = []
        normalized_outputs = []
        json_valid_count = 0

        for i in range(n):
            output = call_openai(prompt)
            raw_outputs.append(output)
            normalized_outputs.append(normalize_text(output))

            if is_valid_json(output):
                json_valid_count += 1

            print(f"\n--- Run {i+1} ---")
            print(output)

        raw_counts = Counter(raw_outputs)
        norm_counts = Counter(normalized_outputs)

        top_raw_count = raw_counts.most_common(1)[0][1]
        top_norm_count = norm_counts.most_common(1)[0][1]

        exact_agreement_rate = top_raw_count / n
        normalized_agreement_rate = top_norm_count / n
        json_valid_rate = json_valid_count / n

        summary = {
            "label": label,
            "runs": n,
            "unique_raw": len(raw_counts),
            "unique_normalized": len(norm_counts),
            "exact_agreement_rate": exact_agreement_rate,
            "normalized_agreement_rate": normalized_agreement_rate,
            "json_valid_rate": json_valid_rate,
            "raw_outputs": raw_outputs
        }

        all_results.append(summary)

        print(f"\nSummary for {label} ({n} runs):")
        print(f"Unique raw outputs: {len(raw_counts)}")
        print(f"Unique normalized outputs: {len(norm_counts)}")
        print(f"Exact agreement rate: {exact_agreement_rate:.2%}")
        print(f"Normalized agreement rate: {normalized_agreement_rate:.2%}")
        print(f"Valid JSON rate: {json_valid_rate:.2%}")

    return all_results

def compare_prompt_versions(results_v1, results_v2):
    print(f"\n\n{'#'*80}")
    print("SIDE-BY-SIDE COMPARISON: VERSION 1 vs VERSION 2")
    print(f"{'#'*80}\n")

    print(f"{'Runs':<10}{'Metric':<30}{'V1':<20}{'V2':<20}")
    print("-"*80)

    for i in range(len(results_v1)):
        r1 = results_v1[i]
        r2 = results_v2[i]
        runs = r1["runs"]

        metrics = [
            ("Unique raw outputs", r1["unique_raw"], r2["unique_raw"]),
            ("Unique normalized outputs", r1["unique_normalized"], r2["unique_normalized"]),
            ("Exact agreement rate", f"{r1['exact_agreement_rate']:.2%}", f"{r2['exact_agreement_rate']:.2%}"),
            ("Normalized agreement rate", f"{r1['normalized_agreement_rate']:.2%}", f"{r2['normalized_agreement_rate']:.2%}"),
            ("Valid JSON rate", f"{r1['json_valid_rate']:.2%}", f"{r2['json_valid_rate']:.2%}")
        ]

        for metric, v1, v2 in metrics:
            print(f"{runs:<10}{metric:<30}{str(v1):<20}{str(v2):<20}")

        print("-"*80)

    print("\nFINAL OBSERVATION:")
    print("If Version 2 has fewer unique outputs, higher agreement, and a higher valid JSON rate,")
    print("then the improved prompt is more consistent and reliable than Version 1.")

# Run both versions
results_v1 = test_prompt(extraction_prompt_v1, "VERSION 1", [5, 10, 15])
results_v2 = test_prompt(extraction_prompt_v2, "VERSION 2", [5, 10, 15])

# Compare side by side
compare_prompt_versions(results_v1, results_v2)
```


Result Analysis:

This result shows that Version 2 is much better and more reliable than Version 1.

Version 1 was inconsistent because the model kept changing the wording and format of the extracted information. Across 5, 10, and 15 runs, it produced many different outputs, with low agreement rates ranging from 20% to 30%, and it never returned valid JSON. Even though the extracted meaning was mostly similar, the structure was unstable, which makes it harder to use in an automated system.

Version 2, on the other hand, was perfectly consistent. In all 5, 10, and 15 runs, it produced the exact same JSON output every time. That gave it 100% agreement and 100% valid JSON rate across all tests.

In short:
The improved prompt worked because it added:

clear instructions
a fixed JSON format
strong constraints

```python
from openai import OpenAI
import json
import re
from collections import defaultdict, Counter
from statistics import mean, stdev
import time

client = OpenAI()

def call_openai(prompt):
    """Call OpenAI API with the provided prompt."""
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content.strip()


def build_extraction_prompt(feedback_text):
    """Build a prompt for structured data extraction from feedback."""
    return f"""
You are a strict data extraction assistant.

Task: Extract structured information from the customer feedback below.

Process:
1. Read the feedback carefully.
2. Identify each required field one by one.
3. Check that each value is explicitly supported by the text.
4. If a field is missing, return "N/A".
5. Determine overall sentiment based only on the feedback.

Output format:
Return ONLY valid JSON with these exact keys:
{{
  "order_id": "",
  "order_date": "",
  "delivery_status": "",
  "packaging_status": "",
  "overall_sentiment": ""
}}

Rules:
- Use only information explicitly stated in the feedback.
- Do not invent, guess, or infer missing details.
- If a field is not mentioned, use "N/A".
- Do not add explanations, notes, or extra text.
- Do not include markdown or code fences.
- Keep each value short and precise.
- "overall_sentiment" must be exactly one of: Positive, Negative, Neutral.

Examples:

Feedback: "My order #99887 arrived on time and in perfect condition."
Output:
{{
  "order_id": "99887",
  "order_date": "N/A",
  "delivery_status": "On time",
  "packaging_status": "Perfect condition",
  "overall_sentiment": "Positive"
}}

Feedback: "Item #55501 was ordered on April 2nd. It never arrived."
Output:
{{
  "order_id": "55501",
  "order_date": "April 2nd",
  "delivery_status": "Not delivered",
  "packaging_status": "N/A",
  "overall_sentiment": "Negative"
}}

Now extract from this feedback:
"{feedback_text}"
"""


TASK_VARIATIONS = [
    ("1", "I ordered item #12345 on March 15th. The delivery was fast but the packaging was damaged.", "Neutral"),
    ("2", "My item #67890 arrived late and the box was crushed.", "Negative"),
    ("3", "I placed my order last week. Delivery was quick, but the packaging was torn.", "Neutral"),
    ("4", "Order #22222 arrived on April 3rd. Delivery was fast and the packaging was in perfect condition.", "Positive"),
    ("5", "I ordered item #33333 on May 1st. It arrived late and the package was badly damaged.", "Negative"),
    ("6", "Order #44444 was placed on June 10th and delivered on June 14th.", "Neutral"),
    ("7", "I ordered item #55555 on July 8th. Delivery was very fast, but the packaging was damaged.", "Neutral"),
    ("8", "Got my order #88888 on March 9th. Shipping was super quick, but the box looked pretty messed up.", "Neutral"),
    ("9", "I was really excited to receive my purchase. I ordered item #99999 on February 14th, and although the delivery was fast, I was disappointed because the packaging was damaged when it arrived.", "Negative"),
    ("10", "Order #77777 was placed on August 2nd. Delivery was delayed, the packaging was damaged, and one item was missing.", "Negative")
]


def is_valid_json(text):
    """Check if text is valid JSON."""
    try:
        json.loads(text)
        return True
    except json.JSONDecodeError:
        return False


def run_all_variations_multiple_times(variations, num_runs=15):
    """Run extraction multiple times for each variation."""
    all_results = {}
    
    print("\n" + "#" * 80)
    print(f"RUNNING EXTRACTION {num_runs} TIMES PER VARIATION")
    print("#" * 80)
    
    total_variations = len(variations)
    
    for var_id, (id_str, feedback, expected_sentiment) in enumerate(variations, start=1):
        print(f"\n{'=' * 80}")
        print(f"Variation {id_str}: Running {num_runs} times...")
        print(f"Input: {feedback}")
        print(f"Expected Sentiment: {expected_sentiment}")
        print(f"{'=' * 80}")
        
        variation_results = []
        
        for run in range(1, num_runs + 1):
            prompt = build_extraction_prompt(feedback)
            output = call_openai(prompt)
            
            valid_json = is_valid_json(output)
            parsed_output = json.loads(output) if valid_json else None
            
            result = {
                "run": run,
                "output": output,
                "valid_json": valid_json,
                "parsed_output": parsed_output,
                "sentiment": parsed_output.get("overall_sentiment") if parsed_output else None,
                "order_id": parsed_output.get("order_id") if parsed_output else None,
                "delivery_status": parsed_output.get("delivery_status") if parsed_output else None,
                "packaging_status": parsed_output.get("packaging_status") if parsed_output else None,
            }
            
            variation_results.append(result)
            
            print(f"  Run {run:2d}: {result['sentiment']:8s} | Valid JSON: {result['valid_json']}")
            time.sleep(0.1)  # Small delay to avoid rate limiting
        
        all_results[id_str] = {
            "feedback": feedback,
            "expected_sentiment": expected_sentiment,
            "runs": variation_results
        }
    
    return all_results


def analyze_consistency(results_by_variation):
    """Analyze consistency of extraction across runs."""
    consistency_report = {}
    
    for var_id, var_data in results_by_variation.items():
        runs = var_data["runs"]
        
        # JSON validity
        valid_count = sum(1 for r in runs if r["valid_json"])
        
        # Sentiment consistency
        sentiments = [r["sentiment"] for r in runs if r["sentiment"]]
        sentiment_counter = Counter(sentiments)
        most_common_sentiment = sentiment_counter.most_common(1)[0][0] if sentiment_counter else None
        sentiment_consistency = (sentiment_counter[most_common_sentiment] / len(sentiments) * 100) if sentiments else 0
        
        # Order ID consistency
        order_ids = [r["order_id"] for r in runs if r["order_id"]]
        order_id_counter = Counter(order_ids)
        order_id_consistency = (order_id_counter.most_common(1)[0][1] / len(order_ids) * 100) if order_ids else 0
        
        consistency_report[var_id] = {
            "json_validity": f"{valid_count}/{len(runs)}",
            "json_validity_percent": valid_count / len(runs) * 100,
            "sentiment_distribution": dict(sentiment_counter),
            "sentiment_consistency": sentiment_consistency,
            "most_common_sentiment": most_common_sentiment,
            "expected_sentiment": var_data["expected_sentiment"],
            "sentiment_accuracy": most_common_sentiment == var_data["expected_sentiment"],
            "order_id_consistency": order_id_consistency,
            "unique_sentiments": len(sentiment_counter)
        }
    
    return consistency_report


def generate_comprehensive_report(results_by_variation):
    """Generate comprehensive comparison report."""
    
    consistency = analyze_consistency(results_by_variation)
    
    print("\n\n" + "#" * 100)
    print("COMPREHENSIVE COMPARISON REPORT - VERSION 3 EXTRACTION PROMPT")
    print("#" * 100)
    
    # Overall Statistics
    print("\n" + "=" * 100)
    print("OVERALL STATISTICS")
    print("=" * 100)
    
    # Count valid JSONs properly
    total_valid = 0
    for var_id, var_data in results_by_variation.items():
        total_valid += sum(1 for r in var_data["runs"] if r["valid_json"])
    
    total_runs = sum(len(results_by_variation[var_id]["runs"]) for var_id in results_by_variation.keys())
    
    print(f"Total Runs: {total_runs}")
    print(f"Total Valid JSON: {total_valid}/{total_runs} ({total_valid/total_runs*100:.1f}%)")
    print(f"Total Variations: {len(consistency)}")
    
    # Sentiment Accuracy
    correct_sentiments = sum(1 for c in consistency.values() if c["sentiment_accuracy"])
    print(f"Sentiment Accuracy: {correct_sentiments}/{len(consistency)} ({correct_sentiments/len(consistency)*100:.1f}%)")
    
    # Sentiment Consistency
    avg_sentiment_consistency = mean(c["sentiment_consistency"] for c in consistency.values())
    print(f"Average Sentiment Consistency: {avg_sentiment_consistency:.1f}%")
    
    # Detailed Variation Analysis
    print("\n" + "=" * 100)
    print("DETAILED VARIATION ANALYSIS")
    print("=" * 100)
    
    for var_id in sorted(consistency.keys(), key=lambda x: int(x)):
        c = consistency[var_id]
        var_data = results_by_variation[var_id]
        
        print(f"\n{'─' * 100}")
        print(f"Variation {var_id}")
        print(f"{'─' * 100}")
        print(f"Feedback: {var_data['feedback']}")
        print(f"\nExpected Sentiment: {c['expected_sentiment']}")
        print(f"Most Common Sentiment: {c['most_common_sentiment']} ({c['sentiment_consistency']:.1f}% consistency)")
        print(f"Accuracy: {'✓ CORRECT' if c['sentiment_accuracy'] else '✗ INCORRECT'}")
        print(f"JSON Validity: {c['json_validity']} ({c['json_validity_percent']:.1f}%)")
        print(f"Sentiment Distribution: {c['sentiment_distribution']}")
        print(f"Order ID Consistency: {c['order_id_consistency']:.1f}%")
        print(f"Unique Sentiments Generated: {c['unique_sentiments']}")
    
    # Improvement Metrics
    print("\n\n" + "=" * 100)
    print("IMPROVEMENT METRICS")
    print("=" * 100)
    
    # Consistency variance
    sentimentConsistencies = [c["sentiment_consistency"] for c in consistency.values()]
    consistency_variance = stdev(sentimentConsistencies) if len(sentimentConsistencies) > 1 else 0
    
    print(f"\nConsistency Metrics:")
    print(f"  - Highest Consistency: {max(sentimentConsistencies):.1f}%")
    print(f"  - Lowest Consistency: {min(sentimentConsistencies):.1f}%")
    print(f"  - Standard Deviation: {consistency_variance:.1f}%")
    
    # Error Analysis
    incorrect_vars = [var_id for var_id, c in consistency.items() if not c["sentiment_accuracy"]]
    if incorrect_vars:
        print(f"\nIncorrect Sentiment Classifications:")
        for var_id in sorted(incorrect_vars, key=lambda x: int(x)):
            c = consistency[var_id]
            print(f"  - Variation {var_id}: Expected {c['expected_sentiment']}, Got {c['most_common_sentiment']}")
    else:
        print(f"\n✓ All sentiment classifications are CORRECT!")
    
    # Stability Report
    print(f"\nStability Report:")
    stable_vars = sum(1 for c in consistency.values() if c["sentiment_consistency"] >= 80)
    unstable_vars = sum(1 for c in consistency.values() if c["sentiment_consistency"] < 80)
    print(f"  - Stable (≥80% consistency): {stable_vars}/{len(consistency)}")
    print(f"  - Unstable (<80% consistency): {unstable_vars}/{len(consistency)}")
    
    if unstable_vars > 0:
        print(f"\n  Unstable Variations:")
        for var_id in sorted(consistency.keys(), key=lambda x: int(x)):
            c = consistency[var_id]
            if c["sentiment_consistency"] < 80:
                print(f"    - Variation {var_id}: {c['sentiment_consistency']:.1f}% (Distribution: {c['sentiment_distribution']})")
    
    # Field Consistency
    print(f"\nField Consistency Analysis:")
    field_stats = defaultdict(lambda: {"consistent": 0, "inconsistent": 0})
    
    for var_id, var_data in results_by_variation.items():
        runs = var_data["runs"]
        valid_runs = [r for r in runs if r["valid_json"]]
        
        if valid_runs:
            order_ids = [r["order_id"] for r in valid_runs]
            delivery_statuses = [r["delivery_status"] for r in valid_runs]
            packaging_statuses = [r["packaging_status"] for r in valid_runs]
            
            order_id_counter = Counter(order_ids)
            delivery_counter = Counter(delivery_statuses)
            packaging_counter = Counter(packaging_statuses)
            
            # Check if single most common value appears in majority
            if order_id_counter and order_id_counter.most_common(1)[0][1] > len(valid_runs) / 2:
                field_stats["order_id"]["consistent"] += 1
            else:
                field_stats["order_id"]["inconsistent"] += 1
            
            if delivery_counter and delivery_counter.most_common(1)[0][1] > len(valid_runs) / 2:
                field_stats["delivery_status"]["consistent"] += 1
            else:
                field_stats["delivery_status"]["inconsistent"] += 1
            
            if packaging_counter and packaging_counter.most_common(1)[0][1] > len(valid_runs) / 2:
                field_stats["packaging_status"]["consistent"] += 1
            else:
                field_stats["packaging_status"]["inconsistent"] += 1
    
    for field, stats in field_stats.items():
        total = stats["consistent"] + stats["inconsistent"]
        consistency_pct = stats["consistent"] / total * 100 if total > 0 else 0
        print(f"  - {field}: {stats['consistent']}/{total} consistent ({consistency_pct:.1f}%)")
    
    # Recommendations
    print("\n" + "=" * 100)
    print("RECOMMENDATIONS")
    print("=" * 100)
    
    if correct_sentiments == len(consistency):
        print("✓ Sentiment classification is highly accurate. No changes needed for sentiment extraction.")
    else:
        print("✗ Sentiment classification needs improvement. Consider:")
        print("  - Adding more explicit sentiment examples")
        print("  - Weighting negative factors (damage, delays) more heavily")
        print("  - Using compound sentiment rules")
    
    if avg_sentiment_consistency >= 90:
        print("✓ Extraction is highly consistent across runs. Model is stable.")
    elif avg_sentiment_consistency >= 70:
        print("⚠ Extraction consistency is moderate. Consider:")
        print("  - Adding more specific extraction rules")
        print("  - Including negative examples in the prompt")
    else:
        print("✗ Extraction consistency is low. Consider:")
        print("  - Restructuring the prompt")
        print("  - Adding Chain-of-Thought more explicitly")
        print("  - Using a more deterministic extraction method")
    
    if unstable_vars > 0:
        all_problematic = set(incorrect_vars)
        for var_id, c in consistency.items():
            if c["sentiment_consistency"] < 80:
                all_problematic.add(var_id)
        
        if all_problematic:
            print(f"⚠ {len(all_problematic)} variations need attention. Review and adjust:")
            for var_id in sorted(all_problematic, key=lambda x: int(x)):
                print(f"  - Variation {var_id}")
    
    print("\n" + "#" * 100)


def create_summary_table(results_by_variation):
    """Create a summary table of all results."""
    consistency = analyze_consistency(results_by_variation)
    
    print("\n\n" + "=" * 100)
    print("QUICK REFERENCE TABLE")
    print("=" * 100)
    
    print(f"\n{'Var':<5} {'Expected':<10} {'Got':<10} {'Accuracy':<10} {'Consistency':<15} {'JSON Valid':<12}")
    print("-" * 100)
    
    for var_id in sorted(consistency.keys(), key=lambda x: int(x)):
        c = consistency[var_id]
        accuracy = "✓" if c["sentiment_accuracy"] else "✗"
        consistency_str = f"{c['sentiment_consistency']:.1f}%"
        
        print(f"{var_id:<5} {c['expected_sentiment']:<10} {c['most_common_sentiment']:<10} {accuracy:<10} {consistency_str:<15} {c['json_validity']:<12}")


if __name__ == "__main__":
    # Run all variations 15 times
    results = run_all_variations_multiple_times(TASK_VARIATIONS, num_runs=15)
    
    # Generate comprehensive report
    generate_comprehensive_report(results)
    
    # Generate summary table        
    create_summary_table(results)
```
